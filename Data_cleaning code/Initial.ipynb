```python
import pandas as pd
import numpy as np
```

```python
from pathlib import Path

base = Path.cwd()
data_path = base / 'Raw data' / 'ravenstack_accounts.csv'
if not data_path.exists():
    data_path = base.parent / 'Raw data' / 'ravenstack_accounts.csv'

df = pd.read_csv(data_path)
df.head()
df.tail()
```

```python
df.head()
```

```python
df_copy = df.copy()
```

```python
df_copy.columns = df_copy.columns.str.strip().str.lower().str.replace(' ', '_')
df_copy.columns
```

```python
df_copy.info()
df_copy.dtypes
```

```python
df_copy = df_copy.apply(lambda s: s.str.strip() if s.dtype == "object" else s)

dtype_map = {
    "account_id": "string",
    "account_name": "string",
    "industry": "category",
    "country": "string",
    "signup_date": "datetime64[ns]",
    "referral_source": "category",
    "plan_tier": "category",
    "seats": "int64",
    "is_trial": "bool",
    "churn_flag": "bool",
}
df_copy = df_copy.astype(dtype_map)
df_copy["signup_date"] = pd.to_datetime(df_copy["signup_date"])
df_copy.info()
```

```python
missing_values = ["n/a", "na", "--", "", "null", "None", "NaN", "nan", "Null", "N/A", "NULL"]
df_copy = df_copy.replace(missing_values, np.nan)
```

```python
df_copy.isna().sum().sort_values(ascending=False)
```

```python
df_copy['account_id'] = df_copy['account_id'].str.strip().str.lower().str.replace('-', '')
df_copy['account_id'].unique()
```

```python
df_copy['account_name'] = df_copy['account_name'].str.strip().str.upper().str.replace('_', '')
df_copy['account_name'].sample(20)
```

```python
df_copy.info()
```

```python
df_copy['account_name'] = df_copy['account_name'].str.strip().str.lower().str.replace(' ', '_')
df_copy['account_name'].sample(20)
```
