```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Documents\Data analytics\projects\my-data-project\SaaS project\Raw data\ravenstack_subscriptions.csv")
```

```python
df_copy = df.copy()
df_copy.sample(20)
```

```python
df_copy.shape
```

```python
df_copy.columns = df_copy.columns.str.replace('_', '').str.lower()
df_copy.columns
```

```python
datatype_mapping = {
    "subscriptionid": "string",
    "accountid": "string",
    "startdate": "datetime64[ns]",
    "enddate": "datetime64[ns]",
    "plantier": "string",
    "seats": "int64",
    "mrramount": "float64",
    "arramount": "float64",
    "istrial": "bool",
    "upgradeflag": "bool",
    "downgradeflag": "bool",
    "churnflag": "bool",
    "billingfrequency": "string",
    "autorenewflag": "bool"
}
df_copy = df_copy.astype(datatype_mapping)
df_copy.dtypes
```

```python
df_copy.duplicated().sum()
```

```python
df_copy.isna().sum()
```

```python
df_copy['subscriptionid'] = df_copy['subscriptionid'].str.strip().str.lower().str.replace('-', '')
df_copy['accountid'] = df_copy['accountid'].str.strip().str.lower().str.replace('-', '')
df_copy['plantier'] = df_copy['plantier'].str.strip().str.lower()
df_copy['billingfrequency'] = df_copy['billingfrequency'].str.strip().str.lower()
df_copy.head()
```

```python
none = ['nan', 'null', 'n/a', 'na', '', 'none', 'None', 'N/A', 'NULL', ' NAN', 'NaN', 'NONE', 'Nan']
df_copy = df_copy.replace(none, np.nan)
df_copy.isna().sum()
```

```python
df_copy.to_csv(r'C:\Documents\Data analytics\projects\my-data-project\SaaS project\Analysis\cleaned_csvs\subscriptions_cleaned.csv', index=False)
```
