```python
df_copy.columns = df_copy.columns.str.replace('_', '').str.lower()
df_copy.columns
```

```python
datatype_mapping = {
    'ticketid': 'string',
    'accountid': 'string',
    'submittedat': 'datetime64[ns]',
    'closedat': 'datetime64[ns]',
    'resolutiontimehours': 'float',
    'priority': 'string',
    'firstresponsetimeminutes': 'float',
    'satisfactionscore': 'float',
    'escalationflag': 'bool'
}

df_copy = df_copy.astype(datatype_mapping)
df_copy.dtypes
```

```python
df_copy.sample(20)
```

```python
df_copy['ticketid'] = df_copy['ticketid'].str.strip().str.lower().str.replace('-', '')
df_copy['accountid'] = df_copy['accountid'].str.strip().str.lower().str.replace('-', '')
df_copy['priority'] = df_copy['priority'].str.strip().str.lower()
```

```python
df_copy.sample(20)
```

```python
df_copy['closedat'] = pd.to_datetime(df_copy['closedat'])
df_copy['closedat'] = df_copy['closedat'].dt.date
df_copy.sample(20)
```

```python
# Ensure closedat is datetime
df['closed_at'] = pd.to_datetime(df['closed_at'])

# Reorder the columns
df_copy = df_copy[['ticketid', 'accountid', 'submittedat', 'closedat', 'resolutiontimehours', 'priority', 'firstresponsetimeminutes', 'satisfactionscore', 'escalationflag']]

df_copy.sample(20)
```

```python
none_values = ['None', 'nan', 'null', 'n/a', 'na', '', 'none', 'N/A', 'NULL', ' NAN', 'NaN']
df_copy = df_copy.replace(none_values, np.nan)
```
