```python
df_copy.columns = df_copy.columns.str.replace('_', '').str.lower()
df_copy.columns
```

```python
data_type_mapping = {
    'usageid': 'string',
    'subscriptionid': 'string',
    'usagedate': 'datetime64[ns]',
    'featurename': 'string',
    'usagecount': 'int64',
    'usagedurationsecs': 'int64',
    'errorcount': 'int64',
    'isbetafeature': 'bool'}

df_copy = df_copy.astype(data_type_mapping)
```

```python
df_copy['usageid'] = df_copy['usageid'].str.strip().str.lower().str.replace('-', '') 
df_copy['subscriptionid'] = df_copy['subscriptionid'].str.strip().str.lower().str.replace('-', '')
df_copy['featurename'] = df_copy['featurename'].str.strip().str.lower().str.replace('_', '')
```
