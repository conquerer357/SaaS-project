```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Documents\Data analytics\projects\my-data-project\SaaS project\Raw data\ravenstack_churn_events.csv")
```

```python
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
df_copy.shape
```

```python
df_copy.info()
```

```python
df_copy.isna().sum()
```

```python
df_copy.columns = df_copy.columns.str.lower().str.replace(' ', '_')
```

```python
none_values = ['nan', 'null', 'n/a', 'na', '', 'none', 'None', 'N/A', 'NULL', ' NAN', 'NaN']

df_copy = df_copy.replace(none_values, np.nan)
df_copy.isna().sum()
```

```python
DTYPE_MAP = {
    "churn_event_id": "string",
    "account_id": "string",
    "churn_date": "datetime64[ns]",
    "reason_code": "string",
    "refund_amount_usd": "float64",
    "preceding_upgrade_flag": "bool",
    "preceding_downgrade_flag": "bool",
    "is_reactivation": "bool",
    "feedback_text": "string",
}
df_copy = df_copy.astype(DTYPE_MAP)
df_copy.info()
```

```python
df_copy['churn_date'] = pd.to_datetime(df_copy['churn_date'])
df_copy.isna().sum()
df_copy.shape
```

```python
df_copy['churn_event_id'] = df_copy['churn_event_id'].str.strip().str.lower().str.replace('-', '')
df_copy['churn_event_id'].sample(5)
```

```python
df_copy['account_id'] = df_copy['account_id'].str.strip().str.lower().str.replace('-', '')
df_copy['account_id'].sample(5)
```

```python
df_copy['reason_code'] = df_copy['reason_code'].str.strip().str.lower()
df_copy['reason_code'].sample(15)
```

```python
df_copy['refund_amount_usd'].sample(20)
```

```python
df_copy['feedback_text'].sample(20)
```

```python
df_copy['feedback_text'].isna().sum()
```

```python
df_copy.to_csv('C:\Documents\Data analytics\projects\my-data-project\SaaS project\Analysis\cleaned_csvs\churn_events_cleaned_2.csv', index=False)
```
